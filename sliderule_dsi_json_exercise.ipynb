## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### 1.  Find the 10 countries with most projects
We understand that each of data represents a project.  To answer this question, the plan is to
1.  Extract all rows with column "countryname" which the country name is the easiest to understand.
2.  Count the number of rows for each of the country.  Here we show two methods of counting by using Counter in Python Collections and directly using dataframe count and that sort the counts in descending order.  Both approaches achieved the same result.
3.  List out the top 10 countries with the highest counts.


First have an initial idea of the column "countryname" by display the first 10 rows.  This display of country name is the most easiest to understand.

```python
country_count_df = json_world_bank['countryname']
```

```python
country_count_df.head(10)
```

```python
from collections import Counter
```

In order to successfully use Counter, convert the Pandas Series country_count_df into a Numpy array.  Then apply Counter and find out the top most counted countries. The result could be readily generated as below, First is the country name following the number of projects.

```python
Counter(country_count_df.values).most_common(10)
```

Here we demonstrate another method on counting number of projects for each country.  The procedure is as below:
1. Exract columns "countryname" and "project name" in json_world_bank dataframe
2. Group the extracted dataframe columns by "countryname"
3. Count the number of projects for each grouped country.
4. Sort the grouped and extracted dataframe by 'project_name' in descending order and display on the top 10 rows.
5. Rename the column index of 'project_name' to 'no_of_projects'

```python
project_count_by_country = json_world_bank[['countryname', 'project_name']].groupby('countryname').count().sort_values(['project_name'], ascending=False).head(10)
```

```python
project_count_by_country.columns = ['no_of_projects']
```

The result is readily shown as below.  You can observe that the result is the same as using Collections Counter:

Result by Counter method previously:

```python
project_count_by_country
```

The goal is to count the number project themes in the column "mjtheme_namecode".  The procedure is as below:

### 3. In 2, above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

First, we analyze the project type code and project type name relationship with the following procedure:
1. Create an empty data frame project_type_df
2. Loop over each row of column "mjtheme_namecode" and extract the list.
3. Loop over the list of each row, extract each the dictionary item.  Convert the dictionary item into a dataframe with the same format as project_type_df.  Append to project_type_df.

Therefore, project_type_df is finally a dataframe containing the full list of project types code and names breaking down the original list and dictionaries.

```python
project_type_df = pd.DataFrame(columns=['code', 'name'], index=['code'])
for item in json_world_bank['mjtheme_namecode']:
   for list_item in item:
    project_type_df = pd.concat([project_type_df, pd.DataFrame(list_item, columns=['code', 'name'], index = ['code'])]).dropna()
```

Rename the column names of project_type_df.

```python
project_type_df.columns = ['project_type_code', 'project_type_name']
```

Obtain the unique value on project_type_name for each project_type_code.  Sort according to project_type_code.

```python
project_type_with_space_df = project_type_df.drop_duplicates().sort_values('project_type_code')
```

We observed:
1. Project type code and project_type_name is a one-to-one mapping.
2. Null string value of project_type_name occurs in all project_type_code values.

The next step is to update the null string value of project_type_name following its project_type_code.

```python
project_type_with_space_df
```

The aim of below procedure is to create a dataframe with project_type_code mapping project_type_name for later referal of value update.  Reuse dataframe project_type_df to filter out those rows with null string value in project_type_name.

```python
project_type_df = project_type_with_space_df[project_type_with_space_df['project_type_name']!='']
```

```python
project_type_df
```

1. Loop over the column "mjtheme_namecode" in json_world_bank dataframe to extract every row of the list.
2. Loop over the extracted list.  Extract the dictionary.  If the 'name' of the dictionary is null string value, assign the project name value which refers to the project_type_df dataframe.

```python
for item in json_world_bank['mjtheme_namecode']:
   for list_item in item:
    if list_item['name'] == '':
        if list_item['code'] == '1':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='1'].project_type_name.to_string(index=None)
        elif list_item['code'] == '2':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='2'].project_type_name.to_string(index=None)
        elif list_item['code'] == '3':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='3'].project_type_name.to_string(index=None)
        elif list_item['code'] == '4':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='4'].project_type_name.to_string(index=None)
        elif list_item['code'] == '5':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='5'].project_type_name.to_string(index=None)
        elif list_item['code'] == '6':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='6'].project_type_name.to_string(index=None)
        elif list_item['code'] == '7':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='7'].project_type_name.to_string(index=None)
        elif list_item['code'] == '8':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='8'].project_type_name.to_string(index=None)
        elif list_item['code'] == '9':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='9'].project_type_name.to_string(index=None)
        elif list_item['code'] == '10':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='10'].project_type_name.to_string(index=None)
        elif list_item['code'] == '11':
            list_item['name'] = project_type_df[project_type_df['project_type_code']=='11'].project_type_name.to_string(index=None)
```
